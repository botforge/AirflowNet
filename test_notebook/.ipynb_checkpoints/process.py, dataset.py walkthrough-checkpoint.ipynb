### def__main__()

```python
sample = test_img
```

```python
print(metadata)
plt.imshow(sample)
```

```python
size = (800, 800)
sample.thumbnail(size, Image.NEAREST)
plt.imshow(sample)
print(sample.size)
```

```python
tensor = transforms.ToTensor()(sample)
tensor.size() #channels, Height, Width
tensor[0, 1, 1] + tensor[1, 1, 1] + tensor[2, 1, 1]
```

```python
#Make Airfoil Tensor 
flattened = torch.sum(tensor, dim = 0) #Add all the three channels
print(flattened.size())
print(flattened[1, 1])
assert(tensor[0, 1, 1] + tensor[1, 1, 1] + tensor[2, 1, 1] == flattened[1, 1])
```

```python
airfoil_mask = torch.where(flattened == 0, flattened, torch.ones_like(flattened)) #put ones wherever the airfoil is
print(airfoil_mask.size())
plt.imshow(airfoil_mask.numpy(), cmap='gray')
plt.colorbar()
```

```python
#Make SDF tensor
binmask = airfoil_mask.numpy()
print(binmask.dtype)
```

```python
binmask = binmask.astype(int)
```

```python
plt.imshow(binmask, cmap='gray')
plt.colorbar()
print(binmask.dtype)
```

```python
posfoil = scipy.ndimage.morphology.distance_transform_edt(binmask)
plt.imshow(posfoil)
plt.colorbar()
```

```python
#testing np.where
a = np.array([1, 2, -1, -3, 1, 2])
b = np.array([1, 2, 3, 4, 5])
print(np.where(a < 0))
print(np.where(b < 0))
```

```python
binmaskcomp = 1 - binmask
plt.imshow(binmaskcomp, cmap='gray')
plt.colorbar()
```

```python
negfoil = scipy.ndimage.morphology.distance_transform_edt(binmaskcomp) #computes SDF for inside the airfoil
plt.imshow(negfoil)
plt.colorbar()
```

```python
sdf = np.subtract(posfoil, negfoil)
plt.imshow(sdf)
plt.colorbar()
```

```python
plot1 = np.where(sdf < 0)
plt.scatter(-1 * plot1[0], plot1[1]) #its just the inner area of the airfoil that is now negative
```

```python
print(sdf[plot1[0], plot1[1]]) #they're not all straight negative one
```

```python
sdf_mask = torch.tensor(sdf).float()
sdf_mask = (sdf_mask / 500) + 0.5
```

```python
#Make pressure tensor
pressure_range = (-1000, 1000)
range_diff = pressure_range[1] - pressure_range[0] 
range_increment = range_diff / 256.0
pressure_mask = (tensor[1, :, :] - 0.5) * range_diff + tensor[2, :, :] * range_increment
# Zero out elements within airfoil
pressure_mask = pressure_mask * airfoil_mask
plt.imshow(pressure_mask.numpy())
plt.colorbar()
```

```python
pressure_mask_numpy = pressure_mask.numpy()
fig = plt.figure()
wh1 = np.where(pressure_mask_numpy < 0) 
plt.scatter(-1*wh1[0], wh1[1])
```
